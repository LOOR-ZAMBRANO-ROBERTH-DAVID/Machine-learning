```python
import pandas as pd
```

```python
datos = pd.read_csv('./datos/musica.csv')
datos
```

Ordenar por nombre de la canción

```python
columnas_seleccionadas = datos[['name', 'artists', 'energy']]
ordernar_por_nombre = columnas_seleccionadas.sort_values(by='name', ascending=True)
ordernar_por_nombre_20 = ordernar_por_nombre.head(20) #Mostrar los primeros 20 registros
ordernar_por_nombre_20
```

Ordenar por el Artista

```python
columnas_seleccionadas = datos[['artists', 'name', 'energy']]
ordernar_por_artista = columnas_seleccionadas.sort_values(by='artists', ascending=True)
ordernar_por_artista_20 = ordernar_por_artista.head(20) #Mostrar los primeros 20 registros
ordernar_por_artista_20
```

Ordenar por Energía


```python
columnas_seleccionadas = datos[['energy', 'name', 'artists']]
ordernar_por_energia = columnas_seleccionadas.sort_values(by='energy', ascending=True)
ordernar_por_energia_20 = ordernar_por_energia.head(20) #Mostrar los primeros 20 registros
ordernar_por_energia_20
```
